```python
# -*- coding: utf-8 -*-
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage

img = cv2.imread("C:\\Users\\56915\\Desktop\\Eiffel_Tower.jpg",cv2.IMREAD_UNCHANGED)

#draw keypoint
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create(0,3,0.31,10,1.6)
kp = sift.detect(gray,None)
img = cv2.drawKeypoints(gray,kp,img)


#rotate 0,45,90
rotate45 = ndimage.rotate(img,-45)
gray45 = cv2.cvtColor(rotate45,cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create(0,3,0.131,10,1.6)
kp3 = sift.detect(gray45,None)
img45 = cv2.drawKeypoints(gray45,kp3,rotate45)

rotate90 = ndimage.rotate(img,-90)
gray90 = cv2.cvtColor(rotate90,cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create(0,3,0.31,10,1.6)
kp4 = sift.detect(gray90,None)
img90 = cv2.drawKeypoints(gray90,kp4,rotate90)

#matches
kp1,des1 = sift.detectAndCompute(img,None)
kp2,des2 = sift.detectAndCompute(img,None)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)

good = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good.append([m])
        
img3 = cv2.drawMatchesKnn(img,kp1,img,kp2,good,None,flags=2)
cv2.imshow("img3",img3)


cv2.waitKey()
cv2.destroyAllWindows()
```
