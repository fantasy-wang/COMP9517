```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.listdir("./")
```

```python
PATH_TO_IMAGE = [os.path.join('./original_retinal_images', i) for i in os.listdir("./original_retinal_images")]
PATH_TO_OD = [os.path.join('./target_images', i) for i in os.listdir("./target_images")]
```

```python
# Load images
imgs = [cv2.imread(PATH, 0) for PATH in PATH_TO_IMAGE] #all the images are 2848*4288
imgs_OD = [cv2.imread(PATH, 0) for PATH in PATH_TO_OD] #all the images are 2848*4288
```

```python
imgs_rgb = [cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB) for PATH in PATH_TO_IMAGE] #all the images are 2848*4288
imgs_OD_rgb = [cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB) for PATH in PATH_TO_OD] #all the images are 2848*4288
```

```python
# For convenience, we convert the imgs to 800*800 to save time
img_resized = [cv2.resize(img, (800, 800)) for img in imgs]
img_OD_resized = [cv2.resize(img, (800, 800)) for img in imgs_OD]

imgs_OD_rgb_resized = [cv2.resize(img, (800, 800)) for img in imgs_OD_rgb]
```

```python
def cvt_Gray2red(gray_img):
    new = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB)
    new[:, :, 0] = gray_img
    new[:, :, 1] = np.zeros([800, 800])
    new[:, :, 2] = np.zeros([800, 800])
    return new

def get_iou(result, original):
    intersection = ((np.concatenate(th1).astype(np.uint16) + np.concatenate(tar_th1).astype(np.uint16)) == 255+255).sum()
    Union = (np.concatenate(result) == 255).sum() + (np.concatenate(original) == 255).sum() - intersection

    return intersection / Union

IoU_list = []

for i in range(len(img_resized)):
    imgt = img_resized[i]
    tar_img = img_OD_resized[i]
    
    ret, th1 = cv2.threshold(imgt, 150, 255, cv2.THRESH_BINARY)

    new_th1 = cvt_Gray2red(th1)
    
    ret, tar_th1 = cv2.threshold(tar_img, 1, 255, cv2.THRESH_BINARY)

    original_target = imgs_OD_rgb_resized[i]
    titles = ['Original', 'BINARY', 'tar_Original', 'tar_BINARY']
    images = [imgt, new_th1, tar_img, original_target]
    IoU = get_iou(th1, tar_th1)
    IoU_list.append(IoU)
    
    # Using matplotlib to display
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.imshow(images[i], 'gray')
        plt.title(titles[i], fontsize=8)
        plt.xticks([]), plt.yticks([])  
        
    plt.show()
    print(f"The IOU of THRESH_BINARY is： {IoU}")
```

```python
np.mean(IoU_list)
```

```python
import skfuzzy as fuzz

cimage = img_resized

for i in range(5):
    img = cimage[i]
    plt.subplot(121), plt.imshow(img, cmap = 'gray')
    plt.title('Original')
    
    ncenters = 5
    
    #get final fuzzy c-partitioned matrix
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(img, ncenters, 2, error=0.005, maxiter=1000, init=None)
    
    #find the max value of every matrics 
    cluster_membership = np.argmax(u, axis=0)
    
    #get the max pixel index of image 
    x_index, y_index = np.unravel_index(np.argmax(img),img.shape)
    
    #find the correspond pixel of cluster_membership
    highest_pixel = cluster_membership[x_index]
    
    #for img if they are not the highest pixel, set them to 0 which is black
    for j in range(ncenters):
        if j != highest_pixel:
            img[cluster_membership == j] = 0

    plt.subplot(122), plt.imshow(img, cmap = 'gray')
    plt.title('Get rid of low clusters')
    plt.show()
```
